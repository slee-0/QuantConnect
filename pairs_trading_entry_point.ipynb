# Pairs Trading Entry Point

$PnL(x)=x(1-N(x))=\frac{x}{\sqrt{2\pi}} \int_x^\infty e^{-\frac{x^2}{2}}dx $

To find the entry point with the maximum expected profit, graph $\frac{x}{2}\text{erfc}\big(\frac{x}{\sqrt{2}}\big)$, where erfc is 1 - error function.

```python
from scipy.special import erf, erfc
import numpy as np
import math
import matplotlib.pyplot as plt

f = plt.figure(figsize=(12, 5), dpi=80, facecolor='w', edgecolor='k')
x = np.linspace(-10, 10, 100)
plt.plot(x,0.5 * x * erfc(x/math.sqrt(2)), linewidth = 2)
```

A zoomed in graph:

```python
f = plt.figure(figsize=(12, 5), dpi=80, facecolor='w', edgecolor='k')
x = np.linspace(0, 1, 64)
plt.plot(x, 0.5 * x * erfc(x/math.sqrt(2)), linewidth = 2)
```

Finding the maximum:

```python
from scipy.optimize import fmin

def f(x): return 0.5 * x * erfc(x/math.sqrt(2))
max_x = fmin(lambda x: -f(x), 0)
print(max_x)
```

Maximum of $\frac{x}{\sqrt{2\pi}}\int_x^\infty e^{-\frac{x^2}{2}}dx$ occurs when $x\approx 0.7518$.
